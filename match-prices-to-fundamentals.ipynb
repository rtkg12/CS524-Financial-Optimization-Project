```python
def match_prices(file):
    raw = pd.read_csv(file)
    
    stock_symbols = raw['Ticker Symbol']
    prices_matched = []
    # Get stock prices in raw df
    for s in stock_symbols:
        date = raw.loc[raw['Ticker Symbol']==s]['Period Ending'].values[0]
        print(date)
        temp = prices.loc[prices['symbol']==s]
        temp1 = temp.loc[temp['date']<=date]
        temp2 = temp.loc[temp['date']>=date]
        if temp1.empty:
            temp = temp2.head(1)
        else:
            temp = temp1.tail(1)
        print(temp)
        print(date)
        close_price = temp['close'].values[0]
        prices_matched.append(close_price)
        #print(close_price)
    
    raw['Close'] = prices_matched
    raw.to_csv(file, index = None, header=True) #Don't forget to add '.csv' at the end of the path
```

```python
match_prices(fundamental_files[1])
match_prices(fundamental_files[2])
```
