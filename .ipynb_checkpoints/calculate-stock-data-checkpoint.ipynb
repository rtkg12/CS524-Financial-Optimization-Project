```python
import numpy as np
import pandas as pd

a = np.genfromtxt('./nyse/prices-split-adjusted.csv', delimiter=',')
b = pd.read_csv('./nyse/prices-split-adjusted.csv')
```

```python
c = b.loc[b['symbol']=="AAPL"] #filter by company symbol
```

```python
d = c['close'].values #store close values as np array in d

d_shifted = np.roll(d, -1) #shift values left by 1
d_shifted
```

```python
d
```

```python
e = ((d_shifted/d)-1)*100 #calculate daily percent return 
# (day 2 / day1) - 1 ) *100
```

```python
e = e[:-1] #remove last element to counteract shift
e
```

```python
np.average(e) #average expected return
```

```python
def reshape_weekly(array):
    size = array.size
    new_size = size - (size%7) #To make it divisible by 7
    array = array[:new_size] #trim last elements to make it divisible by 7
    return array.reshape(-1,7)
```

```python
f = reshape_weekly(f.reshape(-1,1))
f
```
